```python
from pyspark.sql.functions import col
df = spark.read.csv("/Workspace/Users/othmane.95210@gmail.com/Predicting-Road-Accident-Risk/Data/data_HotallCorr.csv", header=True, inferSchema=True)
display(df)
target = df.select(col("accident_risk"))
data = df.drop("accident_risk")
data_pd = data.toPandas()      
target_pd = target.toPandas()

```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_validate

model = Pipeline([
    ("scaler", StandardScaler()),
    ("regressor", LinearRegression())
])

cv_result=cross_validate(model,data_pd,target_pd,cv=5,scoring="neg_mean_squared_error")
scores=-cv_result["test_score"]
print(f"The MSE is {scores.mean():.3f} +/- {scores.std():.3f}")
   
```

We find a 0.005 % MSE its ok because it's a ai dataset now we need to improve that



```python
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_validate

model = Pipeline([
    ("scaler", StandardScaler()),
    ("regressor", LinearRegression())
])

cv_result=cross_validate(model,data_pd,target_pd,cv=5,scoring="neg_mean_squared_error")
scores=-cv_result["test_score"]
rmse_scores = np.sqrt(scores)
print(f"RMSE : {rmse_scores.mean():.6f} +/- {rmse_scores.std():.6f}")
```
