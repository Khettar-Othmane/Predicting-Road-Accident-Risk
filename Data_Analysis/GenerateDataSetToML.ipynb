```python
import pandas as pd
data = pd.read_csv("/Workspace/Users/othmane.95210@gmail.com/Predicting-Road-Accident-Risk/Data/train.csv")
```

```python

from sklearn.preprocessing import OneHotEncoder

num_cols = data.select_dtypes(include='number').columns
obj_cols = data.select_dtypes(include='object').columns
encoder =OneHotEncoder(sparse_output=False).set_output(transform="pandas")
data_Hotencoded=encoder.fit_transform(data[obj_cols])
data_Hotall=pd.concat([data[num_cols], data_Hotencoded], axis=1)
data_Hotall.head()
```

```python
data_HotallCorr=data_Hotall.drop(['num_lanes','road_type_highway','road_type_rural','road_type_urban','time_of_day_afternoon','time_of_day_evening','time_of_day_morning'], axis=1)
```

```python
data_HotallCorrTime=data_Hotall.drop(['time_of_day_afternoon','time_of_day_evening','time_of_day_morning'], axis=1)
```

```python
data_HotallCorrLane=data_Hotall.drop(['num_lanes'], axis=1)
```

```python
data_HotallCorrRoad=data_Hotall.drop(['road_type_highway','road_type_rural','road_type_urban'], axis=1)
```

```python
data_HotallCorr.to_csv("data_HotallCorr.csv", index=False)
data_HotallCorrTime.to_csv("data_HotallCorrTime.csv", index=False)
data_HotallCorrLane.to_csv("data_HotallCorrLane.csv", index=False)
data_HotallCorrRoad.to_csv("data_HotallCorrRoad.csv", index=False)
```
